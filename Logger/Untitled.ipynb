```python
import pandas as pd    
import re
df = open('Log_file.txt', "r")
df1 = open('log_file_edit.txt',"w")
lines = df.readlines()[13:]

for line in lines:
    line = re.sub('[,\[\]\']','',line)
    line = re.sub('[=-]','',line)
    line = line.strip()
    if len(line)<=20:
        line = line.replace(line,"")
    line = re.sub('[|]',',',line)
    line = re.sub(r'\s*,\s*', ',', line)

    print(line)
    df1.write(line+"\n")
df.close()
df1.close()
```
